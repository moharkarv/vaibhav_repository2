```python
# Splitting data into training and testing

train = pd.DataFrame(df1[0:int(len(df1)*0.70)])
test = pd.DataFrame(df1[int(len(df1)*0.70): int(len(df1))])

print(train.shape)
print(test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(train)
data_training_array
```

```python
data_testing_array = scaler.transform(test)
data_testing_array
```

```python
import numpy as np
```

```python
x_train = []
y_train = [] 
n=30
for i in range(n, data_training_array.shape[0]):
    x_train.append(data_training_array[i-n: i])
    y_train.append(data_training_array[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train) 
```

```python
x_test = []
y_test = []

for i in range(n, data_testing_array.shape[0]):
    x_test.append(data_testing_array[i-n: i])
    y_test.append(data_testing_array[i, 0])

x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.initializers import HeNormal

model = Sequential()

model.add(LSTM(units=100, return_sequences=True,input_shape=(x_train.shape[1], x_train.shape[2]),
               kernel_initializer=HeNormal()))

model.add(LSTM(units=100, return_sequences=False,kernel_initializer=HeNormal()))

model.add(Dropout(0.4))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['MAE'])
model.fit(x_train, y_train, validation_data = (x_test, y_test) ,epochs = 50)
```

```python
past_days = pd.DataFrame(train[-n:])
test_df = pd.DataFrame(test)
```

```python
final_df = past_days.append(test_df, ignore_index = True)
```

```python
input_data = scaler.fit_transform(final_df)
input_data
```

```python
x_test = []
y_test = []
for i in range(n, input_data.shape[0]):
   x_test.append(input_data[i-n: i])
   y_test.append(input_data[i, 0])
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
print(x_test.shape)
print(y_test.shape)
```

```python
# Making predictions

y_pred = model.predict(x_test)

```

```python
a=scaler.scale_
```

```python
scale_factor = 1/a
y_pred = y_pred * scale_factor
y_test = y_test * scale_factor
```

```python
plt.figure(figsize = (12,6))
plt.plot(y_test, 'b', label = "Original Price")
plt.plot(y_pred, 'r', label = "Predicted Price")
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print("Mean absolute error on test set: ", mae)
```
